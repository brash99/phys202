The purpose of this code solve the following problem.  A total amount of positive charge, Q, is split between two charges, q1 and q2, which are separated by some distance, d.  Calculate the values of q1 and q2 that maximimize the electric force between the two charges.  The correct answer is q1=q2=Q/2.  We seek to verify this solution by calculating and plotting the magnitude of the electric force for all possible values of q1 between 0 and Q.

We begin by importing matplotlib for plotting purposes.

```python
import matplotlib.pyplot as plt
```

The next step is to define two lists, q1 and force, which will contain the values of q1 and the magnitude of the electric force.  We will also define another list, q1uC, which will contain the charge in microCoulombs.  We also define some sample values of Q, d, and define Coulomb's constant.

```python
k = 8.99E+09
d = 1.0
qtot = 2.0E-06

q1 = [0 for i in range(0,1000)]
force = [0 for i in range(0,1000)]
q1uC = [0 for i in range(0,1000)]
```

Next, we loop over all possible values of q1 between 0 and Q.  At each step, we calculate q1, q1, and the magnitude of the force.

```python
for i in range(0,1000):
    q1[i] = 1.0*i/999.0*qtot
    q2 = qtot - q1[i]
    force[i] = k*q1[i]*q2/(d*d)
    q1uC[i] = q1[i]*1.0E+06
```

Finally, now that we have the appropriate lists, we can plot the results.  We see that the force is maximized when q1 = Q/2.

```python
fig = plt.figure()
ax1 = fig.add_subplot(111)

ax1.plot(q1uC,force)

plt.xlabel('q1 (uC)')
plt.ylabel('Force (N)')
plt.title("Charge Splitting")
```
