# Electric Field from Continuous Charge Distributions

## Example 1: Electric Field from a Line of Charge

Consider a line of charge with total charge $Q$ and length $L$ along the $x$-axis, from $\frac{-L}{2}$ to $\frac{L}{2}$.  We want to find the electric field at a point $P$ a distance $L$ from the origin along the $x$-axis.

<img src="line_of_charge.png" width="800" height="400">

### Step 1: Divide the line of charge into small pieces of length $dx$.

The charge on each piece is $dq = \frac{Q}{L} dx$.

### Step 2: Find the electric field due to each piece of charge $dq$ at point $P$.

The electric field due to a point charge $dq$ at point $P$ is given by

$$d\vec{E} = k_e \frac{dq}{r^2} \hat{i}$$
    
where $r$ is the distance from $dq$ to $P$ and $\hat{i}$ is the unit vector pointing from $dq$ to $P$.  In this case, $r = L-x$, so:
    
$$d\vec{E} = k_e \frac{dq}{(L-x)^2} \hat{i}$$

where $r$ is the distance from $dq$ to $P$ and $\hat{r}$ is the unit vector pointing from $dq$ to $P$.

### Step 3: Integrate over all pieces of charge to find the total electric field at point $P$.

The total electric field at point $P$ is given by

$$\vec{E} = \int d\vec{E} = \int_{-L/2}^{L/2} k_e \frac{dq}{(L-x)^2} \hat{i}$$

where $dq = \frac{Q}{L} dx$.

$$\vec{E} = \int_{-L/2}^{L/2} k_e \frac{Q}{L} \frac{dx}{(L-x)^2} \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \int_{-L/2}^{L/2} \frac{dx}{(L-x)^2} \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \left[ \frac{1}{L-x} \right]_{-L/2}^{L/2} \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \left[ \frac{1}{L-L/2} - \frac{-1}{L+L/2} \right] \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \left[ \frac{1}{L/2} - \frac{-1}{3L/2} \right] \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \left[ \frac{2}{3L/2} \right] \hat{i}$$

$$\vec{E} = \frac{k_e Q}{L} \left[ \frac{4}{3L} \right] \hat{i}$$

$$\vec{E} = \frac{4}{3} k_e \frac{Q}{L^2} \hat{i}$$

Note that this field is larger than the field from a point charge $Q$ at the origin, which is given by

$$\vec{E} = k_e \frac{Q}{L^2} \hat{i}$$


## Example 2: Electric Field from a Half-Circle of Charge

Consider a half-circle of charge with total charge $Q$ and radius $R$ in the $xy$-plane, centered at the origin.  We want to find the electric field at the origin.

<img src="A2Q8.png" width="700" height="500">

The total length of the half-circle is $L = \pi R$.

### Step 1: Divide the half-circle of charge into small pieces of length $dL = R d\theta$.

The charge on each piece is $dq = \frac{Q}{L} dL = \frac{Q}{\pi R} (R d\theta) = \frac{Q}{\pi} d\theta$.

### Step 2: Find the electric field due to each piece of charge $dq$ at the origin.

The electric field due to a point charge $dq$ at the origin is given by

$$d\vec{E} = k_e \frac{dq}{r^2} \hat{r}$$
    
where $r$ is the distance from $dq$ to the origin and $\hat{r}$ is the unit vector pointing from $dq$ to the origin.  In this case, $r = R$, so:

$$d\vec{E} = k_e \frac{dq}{R^2} \hat{r}$$

Expressing $dq$ in terms of $d\theta$:

$$d\vec{E} = k_e \frac{Q}{\pi R} \frac{R d\theta}{R^2} \hat{r}$$

$$d\vec{E} = k_e \frac{Q}{\pi R} \frac{d\theta}{R} \hat{r}$$

$$d\vec{E} = k_e \frac{Q}{\pi R^2} d\theta \hat{r}$$

Now, expressing $\hat{r}$ in terms of $\hat{i}$ and $\hat{j}$:

$$d\vec{E} = k_e \frac{Q}{\pi R^2} d\theta (\cos \theta \hat{j} + \sin \theta \hat{i})$$

### Step 3: Integrate over all pieces of charge to find the total electric field at the origin.

The total electric field at the origin is given by

$$\vec{E} = \int d\vec{E} = \int_{0}^{\pi} k_e \frac{Q}{\pi R^2} d\theta (\cos \theta \hat{j} + \sin \theta \hat{i})$$

$$\vec{E} = \frac{k_e Q}{\pi R^2} \int_{0}^{\pi} d\theta (\cos \theta \hat{j} + \sin \theta \hat{i})$$

$$\vec{E} = \frac{k_e Q}{\pi R^2} \left[ \sin \theta \hat{j} - \cos \theta \hat{i} \right]_{0}^{\pi}$$

$$\vec{E} = \frac{k_e Q}{\pi R^2} \left[ \sin \pi \hat{j} - \cos \pi \hat{i} - \sin 0 \hat{i} + \cos 0 \hat{j} \right]$$

$$\vec{E} = \frac{k_e Q}{\pi R^2} \left[ 0 \hat{j} - (-1) \hat{i} - 0 \hat{j} + 1 \hat{i} \right]$$

$$\vec{E} = \frac{k_e Q}{\pi R^2} \left[ 2 \hat{i} \right]$$

$$\vec{E} = \frac{2}{\pi} k_e \frac{Q}{R^2} \hat{i}$$

This is also a bit of an interesting result.  The point of interest is at a constant distance, $R$, from the charge.  But, the vector nature of the electric field is complicated, and the resulting total electric field is $smaller$ than the field from a point charge $Q$ at the origin, which is given by

$$\vec{E} = k_e \frac{Q}{R^2} \hat{i}$$
