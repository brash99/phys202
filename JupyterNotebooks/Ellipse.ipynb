# Ellipses, Eccentricity, and All That

The definition of an ellipse is:

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$

where $a$ and $b$ are the semi-major and semi-minor axes, respectively. The eccentricity is defined as: 

$$e = \sqrt{1 - \frac{b^2}{a^2}}$$

The eccentricity is a measure of how "squished" the ellipse is. A circle has $e = 0$, while a line segment has $e = 1$.

The eccentricity can also be defined as:

$$e = \frac{c}{a}$$

where $c$ is the distance from the center of the ellipse to the focus. The eccentricity is also the ratio of the distance between the foci and the length of the major axis.    



```python
import numpy as np
import matplotlib.pyplot as plt

# Draw a plot of an ellipse

e = 0.8
a = 1
b = a * np.sqrt(1 - e**2)

x = np.linspace(-a, a, 100)
y = b * np.sqrt(1 - x**2 / a**2)

plt.plot(x, y)
plt.plot(x, -y)
plt.axis('equal')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Ellipse')
plt.show()
```

# Coordinate Transformations

It is particularly interesting to choose a coordinate syatem where the origin is at one focus of the ellispes. In this case, the equation of the ellipse becomes:

$$\frac{x^2}{a^2} + \frac{y^2}{a^2 - c^2} = 1$$

where $c$ is the distance from the origin to the focus. The eccentricity is now:

$$e = \sqrt{1 - \frac{a^2 - c^2}{a^2}} = \sqrt{1 - \frac{c^2}{a^2}}$$

The equation of the ellipse can be rewritten as:

$$\frac{x^2}{a^2} + \frac{y^2}{a^2 - c^2} = 1$$

## Polar Coordinates

The equation of an ellipse in polar coordinates is:

$$r = \frac{a(1 - e^2)}{1 - e \cos \theta}$$

where $r$ is the distance from the origin and $\theta$ is the angle from the major axis. The eccentricity is:

$$e = \sqrt{1 - \frac{b^2}{a^2}}$$

```python
# Draw a plot of an ellipse in polar coordinates

e = 0.8
a = 1
b = a * np.sqrt(1 - e**2)

theta = np.linspace(0, 2 * np.pi, 100)
r = a * (1 - e**2) / (1 - e * np.cos(theta))

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
ax.plot(theta, r)
ax.set_title('Ellipse in Polar Coordinates')
plt.show()
```

## Why are Orbits Ellipses?

The gravitational force between two objects is:

$$F = \frac{G m_1 m_2}{r^2}$$

where $G$ is the gravitational constant, $m_1$ and $m_2$ are the masses of the two objects, and $r$ is the distance between the two objects. The gravitational force is a central force, which means that it is always directed along the line connecting the two objects.

The gravitational force is a conservative force, which means that the total energy is conserved. The total energy is the sum of the kinetic and potential energies:

$$E = \frac{1}{2} m v^2 - \frac{G m_1 m_2}{r}$$

where $v$ is the speed of the object. The total energy is constant, which means that the kinetic and potential energies can change, but the total energy is always the same.

In order to proceed, we need to write the velocity in terms of $r$ and $\theta$. Referring to the figure below, we can see that the velocity has two components: one in the radial direction and one in the tangential direction. The radial component is is the change in $r$ with time, which is $\frac{dr}{dt} = \dot{r}$. The tangential component is the change in $\theta$ with time, which is $r \omega = r \frac{d\theta}{dt} = r \dot{\theta}$.

![Orbit](Elliptical_Orbit.jpg "Orbit")

Thus, the velocity is:

$$v = \sqrt{\dot{r}^2 + r^2 \dot{\theta}^2}$$

where $\dot{r}$ and $\dot{\theta}$ are the time derivatives of $r$ and $\theta$, respectively. The kinetic energy is:

$$K = \frac{1}{2} m v^2 = \frac{1}{2} m (\dot{r}^2 + r^2 \dot{\theta}^2)$$

The potential energy is:

$$U = - \frac{G m_1 m_2}{r}$$

Thus, finally, the total energy (which is conserved) is:

$$E = \frac{1}{2} m (\dot{r}^2 + r^2 \dot{\theta}^2) - \frac{G m_1 m_2}{r}$$

Now, the angular momentum (which is also conserved!) is:

$$L = I \omega = (m r^2) \dot{\theta}$$

## The Orbit Equation

We begin by making the substitution: $\rho = \frac{1}{r}$. 

We can then write the angular momentum equation as:

$$\theta = \int \frac{L}{m r^2} dt = \int \frac{L}{m} \rho^2 dt = \int \frac{L}{m} \rho^2 \frac{dt}{d\rho} d\rho$$

$$ r = \frac{1}{\rho}$$
$$ \dot{r} = - \frac{1}{\rho^2} \frac{d\rho}{dt}$$

Substituting, we see that:

$$\theta = \int \frac{L}{m \dot{r}} d\rho$$ 

Now, turning to the energy equation, we can also make the substitution: $\rho = \frac{1}{r}$, and also substitute for $\dot{\theta}$ from the angular momentum equation:

After rearranging, we get:

$$\dot{r} = \frac{2E}{m} + 2GM\rho - \frac{L^2}{m^2} \rho^2$$

We are now going to make two non-obvious substitutions:

$$ r_0 = \frac{L^2}{m^2 G M}$$
$$ e^2 = 1 - \frac{2 E r_0}{GMm}$$

Note that both $r_0$ and $e^2$ are constants.  We can now write the equation for $\dot{r}$ as:

$$\dot{r} =  \frac{L}{m} \sqrt{\frac{e^2}{r_0^2} - \left( \rho - \frac{1}{r_0} \right)^2}$$

Now, we can substitute for $\dot{r}$ in the integral equation for $\theta$:

$$\theta = \int \frac{L}{m \dot{r}} d\rho = \int \frac{1}{\sqrt{\frac{e^2}{r_0^2} - \left( \rho - \frac{1}{r_0} \right)^2}} d\rho$$

It might not seem obvious, but this is a fairly well-known integral form. The solution is:

$$\theta = \cos^{-1} \left( \frac{\rho - \frac{1}{r_0}}{\frac{e}{r_0}} \right)$$

which can be rearranged to give:

$$r = \frac{r_0}{1 - e \cos \theta}$$

This is exactly the equation for an ellipse, with the Sun at one focus!  Thus, we have shown that the orbit of a planet around the sun is an ellipse.

By looking at the values of $r$ at $\theta = 0$ and $\theta = \pi$, we can easily show that:

$$r_0 = a(1 - e^2)$$

where $a$ is the semi-major axis of the ellipse.  Thus, we can write the equation for the orbit as:

$$r = \frac{a(1 - e^2)}{1 - e \cos \theta}$$



```python
# Draw the orbit of Mercury

e = 0.206
a = 0.3871 # AU
b = a * np.sqrt(1 - e**2)

theta = np.linspace(0, 2 * np.pi, 100)
r = a * (1 - e**2) / (1 - e * np.cos(theta))

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
ax.plot(theta, r)
ax.set_title('Orbit of Mercury')
plt.show()
```

## Kepler's Laws

Kepler's laws are:

1. The orbit of each planet around the Sun is an ellipse with the Sun at one focus.  We have already shown this, rigorously, above.
2. The line joining the Sun and a planet sweeps out equal areas in equal times as the planet travels around the ellipse. This is a consequence of the conservation of angular momentum.
3. The square of the orbital period of a planet is proportional to the cube of the semi-major axis of its orbit. This is a consequence of the conservation of energy.  The constant of proportionality is the same for all objects orbiting the Sun (or any other object).

Let's consider the third law.  We can write this as:

$$\frac{a^3}{T^2) = k$$ 

where $T$ is the orbital period, $a$ is the semi-major axis, and $k$ is a constant.

Let's consider a circular orbit.  The orbital period is:

$$T = \frac{2 \pi r}{v}$$

where $r$ is the radius of the orbit and $v$ is the speed of the object.  The speed of the object is:

$$v = \sqrt{\frac{G M}{r}}$$

where $M$ is the mass of the object being orbited.  Thus, the orbital period is:

$$T = \frac{2 \pi r}{\sqrt{\frac{G M}{r}}} = 2 \pi \sqrt{\frac{r^3}{G M}}$$

Thus, we can write the third law as:

$$\frac{a^3}{T^2} = \frac{a^3}{4 \pi^2 \frac{r^3}{G M}} = \frac{G M}{4 \pi^2}$$

```python
# Calculate the time dependence of the position of Mercury

e = 0.206
a = 0.3871 # AU
b = a * np.sqrt(1 - e**2)

theta = np.linspace(0, 2 * np.pi, 100)
r = a * (1 - e**2) / (1 - e * np.cos(theta))
v = np.sqrt(1.327e20 * (2 / r - 1 / a)) # m/s

# calculate the time dependence of the position
t = np.linspace(0, 1.0 * 365 * 24 * 3600, 100)
x = r * np.cos(theta)
y = r * np.sin(theta)

x = x * 1.496e11 # convert to meters
y = y * 1.496e11 # convert to meters
vx = v * np.cos(theta)
vy = v * np.sin(theta)
vx = vx * 1.496e11 / (365 * 24 * 3600) # convert to AU/yr
vy = vy * 1.496e11 / (365 * 24 * 3600) # convert to AU/yr

fig, ax = plt.subplots(1, 2, figsize=(12, 6))
ax[0].plot(t, x)
ax[0].set_xlabel('t (s)')
ax[0].set_ylabel('x (m)')
ax[0].set_title('x vs. t')
ax[1].plot(t, y)
ax[1].set_xlabel('t (s)')
ax[1].set_ylabel('y (m)')
ax[1].set_title('y vs. t')
plt.show()
```
