# Ellipses, Eccentricity, and All That

The definition of an ellipse is:

$$\frac{x^2}{a^2} + \frac{y^2}{b^2} = 1$$

where $a$ and $b$ are the semi-major and semi-minor axes, respectively. The eccentricity is defined as: 

$$e = \sqrt{1 - \frac{b^2}{a^2}}$$

The eccentricity is a measure of how "squished" the ellipse is. A circle has $e = 0$, while a line segment has $e = 1$.

The eccentricity can also be defined as:

$$e = \frac{c}{a}$$

where $c$ is the distance from the center of the ellipse to the focus. The eccentricity is also the ratio of the distance between the foci and the length of the major axis.    



```python
import numpy as np
import matplotlib.pyplot as plt

# Draw a plot of an ellipse

e = 0.8
a = 1
b = a * np.sqrt(1 - e**2)

x = np.linspace(-a, a, 100)
y = b * np.sqrt(1 - x**2 / a**2)

plt.plot(x, y)
plt.plot(x, -y)
plt.axis('equal')
plt.show()
```

# Coordinate Transformations

It is particularly interesting to choose a coordinate syatem where the origin is at one focus of the ellispes. In this case, the equation of the ellipse becomes:

$$\frac{x^2}{a^2} + \frac{y^2}{a^2 - c^2} = 1$$

where $c$ is the distance from the origin to the focus. The eccentricity is now:

$$e = \sqrt{1 - \frac{a^2 - c^2}{a^2}} = \sqrt{1 - \frac{c^2}{a^2}}$$

The equation of the ellipse can be rewritten as:

$$\frac{x^2}{a^2} + \frac{y^2}{a^2 - c^2} = 1$$

## Polar Coordinates

The equation of an ellipse in polar coordinates is:

$$r = \frac{a(1 - e^2)}{1 - e \cos \theta}$$

where $r$ is the distance from the origin and $\theta$ is the angle from the major axis. The eccentricity is:

$$e = \sqrt{1 - \frac{b^2}{a^2}}$$

```python
# Draw a plot of an ellipse in polar coordinates

e = 0.8
a = 1
b = a * np.sqrt(1 - e**2)

theta = np.linspace(0, 2 * np.pi, 100)
r = a * (1 - e**2) / (1 - e * np.cos(theta))

fig, ax = plt.subplots(subplot_kw={'projection': 'polar'}, figsize=(8, 8))
ax.plot(theta, r)
plt.show()
```
