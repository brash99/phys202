```python
# part of the magwire package
# calculate magnetic fields arising from electrical current through wires of arbitrary shape
# with the law of Biot-Savart
# written by Michael Wack 2015
# wack@geophysik.uni-muenchen.de
# tested with python 3.4.3
# some basic calculations for testing
#
# edits, mods, updates by E.J. Brash, February 2020

import math
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import plotly.offline as py
import plotly.graph_objs as go

#
# Get local wire and biotsavart classes
#
import os
import sys
sys.path.append(os.getcwd())
import wire
import biotsavart
```

```python
wirelist = []

# simple wire
#wirelist.append(wire.Wire(path=wire.Wire.LinearPath([0.5,0.5,0.0],[0.5,0.5,1.0],pts=200), 
#              discretization_length=0.01, current=100000))
#
#dmin = 0.15 # minimum x,y, or z value of grid
#dmax = 0.85 # maximum x,y, or z value of grid
#sol = biotsavart.BiotSavart(wire=wirelist[0])

# single wire loop
wirelist.append(wire.Wire(path=wire.Wire.SinusoidalCircularPath(radius=0.5,amplitude=0.0,frequency=12.0,pts=200), 
              discretization_length=0.01,current=2000))
dmin = -0.70 # minimum x,y, or z value of grid
dmax = 0.70 # maximum x,y, or z value of grid
sol = biotsavart.BiotSavart(wire=wirelist[0])

# assembly of 10 wire loops
#for i in range(21):
#    wirelist.append(wire.Wire(path=wire.Wire.SinusoidalCircularPath(radius=0.5,amplitude=0.0,frequency=12.0,pts=200), discretization_length=0.01,current=2000).Translate([0.0,0.0,-0.5+i*0.05]))
#sol = biotsavart.BiotSavart(wire=wirelist[0])
#for i in range(1,21):
#    sol.AddWire(wirelist[i])
#dmin = -1.20 # minimum x,y, or z value of grid
#dmax = 1.20 # maximum x,y, or z value of grid

gridsize = 8 # number of grid points in each of x,y, and z
```

```python
resolution = (dmax-dmin)/(gridsize-1)
volume_corner1 = (dmin, dmin, dmin)
volume_corner2 = (dmax+0.001, dmax+0.001, dmax+0.001)

grid = np.mgrid[volume_corner1[0]:volume_corner2[0]:resolution, 
                volume_corner1[1]:volume_corner2[1]:resolution,
                volume_corner1[2]:volume_corner2[2]:resolution]
#print (grid)

# create list of grid points
points = np.vstack(list(map(np.ravel, grid))).T


# calculate B field at given points
B = sol.CalculateB(points=points)
Bx = B[:,0]
By = B[:,1]
Bz = B[:,2]
Babs = np.linalg.norm(B, axis=1)
for i in range(len(Babs)):
    if math.isnan(Babs[i]):
        Babs[i]=0.0
        
Bgrid=(Babs.reshape(gridsize,gridsize,gridsize))
Bxgrid=(Bx.reshape(gridsize,gridsize,gridsize))
Bygrid=(By.reshape(gridsize,gridsize,gridsize))
Bzgrid=(Bz.reshape(gridsize,gridsize,gridsize))
#pointgrid=points.reshape(8,8,8)

#print (Bgrid[:][3][:])
#print (Bgrid[:][4][:])
```

```python
#Draw results
# remove big values close to the wire
cutoff = 4.0
minvalue = 0.0002

B[Babs > cutoff] = [np.nan,np.nan,np.nan]
B[Babs < minvalue] = [np.nan,np.nan,np.nan]

fig = plt.figure()
# 3d quiver
ax = fig.gca(projection='3d')
sol.mpl3d_PlotWires(ax)
ax.quiver(points[:, 0], points[:, 1], points[:, 2], B[:, 0], B[:, 1], B[:, 2], length=0.8)
plt.show()
```

```python
# colormap of the total magnetic field size
# here, the cm.jet is used
cmfunc = lambda x: 255*cm.jet(x)
pl_BrBG = [ [x, "rgb(%d,%d,%d)"%(255*cmfunc(x)[0],255*cmfunc(x)[1],255*cmfunc(x)[2])] for x in np.linspace(0,1,11) ]

def get_the_slice(x,y,z, surfacecolor,  colorscale=pl_BrBG, showscale=False):
    return go.Surface(x=x,# https://plot.ly/python/reference/#surface
                   y=y,
                   z=z,
                   surfacecolor=surfacecolor,
                   colorscale=colorscale,
                   showscale=showscale,
                   name="")

def get_lims_colors(surfacecolor):# color limits for a slice
    return np.min(surfacecolor), np.max(surfacecolor)
```

```python
data = []

for w in wirelist:
    x = (w.discretized_path).T[0]
    y = (w.discretized_path).T[1]
    z = (w.discretized_path).T[2]

    trace = go.Scatter3d(
        x=x, y=y, z=z,
        line=dict(color='#000000',width=4),
        mode="lines",
        name="",
        showlegend=False,
    )
    data.append(trace)
```

```python
x = np.linspace(dmin,dmax, gridsize)
z = np.linspace(dmin,dmax, gridsize)
z,x = np.meshgrid(x,z)
y = np.array([[((dmax+dmin)/2.0) for i in range(len(x))] for j in range(len(z))])

surfcolor_y = (Bgrid[:][int(gridsize/2-1)][:]+Bgrid[:][int(gridsize/2)][:])/2
Bxgrid_y = (Bxgrid[:][int(gridsize/2-1)][:]+Bxgrid[:][int(gridsize/2)][:])/2
Bzgrid_y = (Bzgrid[:][int(gridsize/2-1)][:]+Bzgrid[:][int(gridsize/2)][:])/2

#print((Bgrid[:][3]+Bgrid[:][4][:])/2)
#print (surfcolor_y)
slice_y = get_the_slice(x,y,z, surfcolor_y)
#print (slice_y)

y=np.linspace(dmin,dmax, gridsize)
z=np.linspace(dmin,dmax, gridsize)
z,y=np.meshgrid(y,z)

x = np.array([[((dmax+dmin)/2.0) for i in range(len(y))] for j in range(len(z))])

surfcolor_x = (Bgrid[int(gridsize/2-1)]+Bgrid[int(gridsize/2)])/2
Bygrid_x = (Bygrid[:][int(gridsize/2-1)][:]+Bygrid[:][int(gridsize/2)][:])/2
Bzgrid_x = (Bzgrid[:][int(gridsize/2-1)][:]+Bzgrid[:][int(gridsize/2)][:])/2

#print (surfcolor_x)
slice_x = get_the_slice(x,y,z, surfcolor_x)
#print (slice_x)

data.append(slice_y)
data.append(slice_x)

data.append(trace)
```

```python
Bmin,Bmax = get_lims_colors(surfcolor_y)
slice_x.update(cmin=0.0, cmax=Bmax)
slice_y.update(cmin=0.0, cmax=Bmax, showscale=True,
               colorbar=go.surface.ColorBar(title='B Field [Tesla]'),)

#print (get_lims_colors(surfcolor_y))
#print (get_lims_colors(surfcolor_x))

axis = dict(showbackground=True,
            backgroundcolor="rgb(230, 230,230)",
            gridcolor="rgb(255, 255, 255)",
            zerolinecolor="rgb(255, 255, 255)",
            )

layout = go.Layout(
         width=700,
         height=700,
         scene=go.layout.Scene(xaxis=go.layout.scene.XAxis(axis),
                        yaxis=go.layout.scene.YAxis(axis),
                        zaxis=go.layout.scene.ZAxis(axis, range=[0.9*dmin,1.1*dmax]),
                        aspectratio=dict(x=1, y=1, z=1),
                        ),
        )
```

```python
fig = dict(data=data,layout=layout)
py.iplot(fig)
```

```python
xgrid = np.linspace(dmin,dmax, gridsize)
ygrid = np.linspace(dmin,dmax, gridsize)
zgrid = np.linspace(dmin,dmax, gridsize)

fig, ax = plt.subplots()

# Plot the streamlines with an appropriate colormap and arrow style
color = 2 * np.log(np.hypot(Bygrid_x, Bzgrid_x))
ax.streamplot(ygrid, zgrid, Bygrid_x, Bzgrid_x, color=color, linewidth=1, cmap=plt.cm.inferno,
              density=2, arrowstyle='->', arrowsize=1.5)
```
