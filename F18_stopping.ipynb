### Numerical calculation of the stopping of an F-18

Using Newton's 2$^{nd}$ Law, we determined that:

$ a = - \frac{2T}{m} cos(\theta) = - \frac{2T}{m} \frac{y}{\sqrt{y^2+\Delta x^2}}$

```python
# import the standard numpy and matplotlib libraries
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# set the initial speed and tension in the cable
# the tension was adjusted until the final position came out
# to be 72 ft.

vi = 210.0
T = 6.204E06

# set the x-length of the cable (1/2 distance) and the mass of the F18
dx = 33.0
m = 26000.0
```

```python
# Choose a time-step for the numerical calculation (in seconds)
# Choose a number of time steps ... this was adjusted so that the
# plane would reach zero speed in this amount of time

dt = 0.001
nsteps = 600

# create some arrays to hold the time values, the velocity at each
# time step, the acceleration at each time step, and the position (y)
# at each time step

t = np.zeros(nsteps)
v = np.zeros(nsteps)
a = np.zeros(nsteps)
y = np.zeros(nsteps)

# Set the initial position, initial velocity, and initial acceleration
y[0] = 0.0
v[0] = vi
a[0] = -2.0*T/m*y[0]/np.sqrt(y[0]**2+dx**2)

# Create the array of time values
for i in range(1,nsteps):
    t[i] = t[i-1]+dt

# Calculate the new velocity, position, and acceleration at each
# time step.
for i in range(1,nsteps):
    v[i] = v[i-1]+a[i-1]*dt # v_new = v_old + a*dt
    y[i] = y[i-1]+v[i]*dt # y_new = y_old + v_new*dt
    a[i] = -2.0*T/m*y[i]/np.sqrt(y[i]**2+dx**2) # a_new = f(y_new) as above

    # stop the loop if v[i] < 0 ... i.e. if the plane has stopped.
    # The cable tension was adjusted such that this happens at y = 72 ft.
    if (v[i] < 0.0):
        print ("Final position = ",y[i])
        break

# Plot the velocity vs. time graph
plt.plot(t,v)
```

```python
# Plot the acceleration vs. time graph
plt.plot(t,a)
```

```python
# Plot the position vs. time graph
plt.plot(t,y)
```
